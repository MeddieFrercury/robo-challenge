```python
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
```

```python
def show_image(image_bgr, t=cv2.COLOR_BGR2RGB):
    image_rbg = cv2.cvtColor(image_bgr, t)
    plt.imshow(image_rbg)
    plt.show()
```

```python
import cv2
import rps

image = cv2.imread('test-images/robot.jpg')

location = rps.RobotPositioningSystem.locate_robot(image, rps.WHITE)

print(location)
show_image(image)
```
